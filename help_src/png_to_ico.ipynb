This script is designed to converte a png file to a proper ico file

## Step 2: Png 2 ICO

````python

from PIL import Image
````

````python
# Input PNG file
png_file = "../images/mainanalytics_Logo_m.png"

# Output ICO file
ico_file = "../images/ma.ico"

# Open the PNG
img = Image.open(png_file)

# Save as ICO with multiple sizes for Windows compatibility
#sizes = [(16, 16), (32, 32), (48, 48), (64, 64), (128, 128), (256, 256)]
sizes = [(128, 128), (256, 256)]
img.save(ico_file, format="ICO", sizes=sizes)

print(f"Converted {png_file} to {ico_file}")
````
